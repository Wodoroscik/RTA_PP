```python
#SparkSession — brama do Sparka (odpowiednik połączenia z bazą danych)
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Lab2-Transactions")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("WARN")
print(f"Spark {spark.version} — gotowy")
```

```python
#wczytanie JSONA
df = spark.read.json("transactions_10k.jsonl")

print(f"Liczba rekordów: {df.count()}")
df.printSchema()
```

```python
df.show(10, truncate=False)
```

```python
# Konwersja kolumny stringowej na typ timestamp 
from pyspark.sql.functions import to_timestamp, col

df = df.withColumn("timestamp", to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss"))

df.printSchema()  # timestamp powinien być teraz 'timestamp (nullable = true)'
```

```python
# Liczba transakcji i suma przychodów per sklep
from pyspark.sql.functions import count, sum as _sum, avg, round as _round

store_summary = (
    df.groupBy("store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _round(avg("amount"), 2).alias("srednia_PLN"),
    )
    .orderBy("store")
)
store_summary.show()
```

```python
# Statystyki per kategoria (Zad. 2.2)
from pyspark.sql.functions import min as _min, max as _max

store_summary = (
    df.groupBy("category")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _min("amount").alias("kwota minimalna"),
        _max("amount").alias("kwota maksymalna")
    )
    .orderBy("category")
)
store_summary.show()
```

```python
# Liczba transakcji per godzina (tumbling 1h)
from pyspark.sql.functions import window

hourly = (
    df.groupBy(window("timestamp", "1 hour"))    # okno 1-godzinne
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .orderBy("window")
)
hourly.show(truncate=False)
```

```python
# Kolumna window to struct z polami start i end. Możemy je wyciągnąć żeby ładniej wyświetlić:
(
    hourly
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .show(truncate=False)
)
```

```python
# Okna 30-minutowe per sklep (Zad. 3.2)
thirty_min = (
    df.groupBy(window("timestamp", "30 minutes"), "store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy("window", "store")
)
thirty_min.show(truncate = False)
```

```python
# W której godzinie sklep “Kraków” miał najwyższy przychód? (Zad. 3.3)
# Filtruj najpierw po sklepie, potem zrób okno godzinne, posortuj malejąco po sumie.
from pyspark.sql.functions import desc
krakow_max = (
    df
    .filter(col("store") == "Kraków")
    .groupBy(window("timestamp", "1 hour"))
    .agg(
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy(desc("suma_PLN"))
)
krakow_max.show(truncate = False)
```

```python
# Okno sliding (przesuwne) — przedziały nakładają się:
# Dwa parametry: windowDuration (szerokość okna) i slideDuration (co ile przesuwa się).
sliding = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))  # szerokość 1h, krok 30min
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .orderBy("od")
)
sliding.show(truncate=False)
```

```python
tumbling_rows = (
    df.groupBy(window("timestamp", "1 hour"))
    .agg(count("tx_id"))
    .count()
)
sliding_rows = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))
    .agg(count("tx_id"))
    .count()
)
print(f"Tumbling (1h):          {tumbling_rows} okien")
print(f"Sliding  (1h / 30min):  {sliding_rows} okien")

# Odpowiedz w komentarzu: dlaczego sliding ma więcej wierszy? (Zad. 4.2)
# Sliding ma więcej wierszy ze względu na overlap - okna nachodzą na siebie, a w tumbling są równej długości.
```

```python
# Odpowiedz na pytania w komentarzach (Zad. 5):

# 1. Ile transakcji jest w oknie 09:00–10:00?
#    Sprawdź w wyniku zadania 3.1.
#    ODPOWIEDŹ: 3

# 2. Jaka jest różnica między groupBy("store") a groupBy(window(...), "store")?
#    ODPOWIEDŹ: pierwszy wariant agreguje globalnie (jeden sklep = jeden wiersz), a drugi chronologicznie 
#              (wiele wierszy dla każdego sklepu, po jednym dla każdego okna czasowego)

# 3. W oknie sliding 1h/30min — ile okien zawiera transakcje z godziny 09:30?
#    Wskazówka: narysuj oś czasu.
#    ODPOWIEDŹ: dwa okna, ponieważ 'długość okna' / 'krok' = (60min / 30min = 2)
```

```python
# Znajdź godzinę, w której sklep Gdańsk miał najniższą średnią kwotę transakcji. (PD. 1)
from pyspark.sql.functions import asc
pd_1 = (
    df
    .filter(col("store") == "Gdańsk")
    .groupBy(window("timestamp", "1 hour"))
    .agg(
        _round(avg("amount"), 2).alias("srednia_kwota_transakcji")
    )
    .orderBy(asc("srednia_kwota_transakcji"))
)
pd_1.show(1, truncate = False)
```

```python
# Policz ile transakcji per kategoria było w oknie 09:00–09:30. (PD. 2)
from pyspark.sql import functions as F

pd_2 = (
    df
    .groupBy(
        F.window("timestamp", "30 minutes"),
        "category"
    )
    .agg(
        F.count("tx_id").alias("liczba_tx")
    )
    .filter(F.col("window.start").cast("string").contains("09:00:00"))
    .orderBy("liczba_tx", ascending=False)
)

pd_2.show(truncate=False)
```

```python
# Zrób okno 15-minutowe i sprawdź w której ćwierćgodzinie był szczyt transakcji (łącznie dla wszystkich sklepów). (PD. 3)
pd_3 = (
    df
    .groupBy(F.window("timestamp", "15 minutes"))
    .count()
    .orderBy(F.desc("count"))
)
pd_3.show(1, truncate=False)
```
